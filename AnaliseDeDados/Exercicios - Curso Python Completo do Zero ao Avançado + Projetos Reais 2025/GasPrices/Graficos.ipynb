```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gas_prices.csv')
df
```

```python
plt.plot(df['Year'], df['Australia'], '.-',  label = 'Australia')
plt.plot(df['Year'], df['Italy'], '.-', label = 'Italy')
plt.plot(df['Year'], df['USA'],'.-', label = 'USA')
plt.plot(df['Year'], df['Japan'] ,'.-', label = 'Japan')

plt.xticks(df['Year'][::2])

plt.xlabel('Years')
plt.ylabel('US$')
plt.legend()
plt.savefig('countries.png')
plt.show()
```
