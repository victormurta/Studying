```python
import pandas as pd
import re #bilbio das expressões regulares, facilita para fazer condições melhores
```

```python
df = pd.read_csv('house.csv')
```

```python
#é possível criar variáveis com filtros separados
filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] >= 500000)]
filter1
```

```python
filter2 = df.loc[df['Address'].str.contains('Turner st|Turner Rd', flags=re.I)]
filter2
```

```python
filter3 = df.loc[df['Address'].str.contains('^59', flags=re.I) | df['Price'] <= 800000 ]
filter3
```

```python
#filtros sendo utilizados para fazer alterações em elementos específicos
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronald'
df
```

```python
df.loc[df['SellerG'] == 'Ronald', 'Method'] = 'Pending'
df
```

```python
#Agrupamento
df.groupby(['SellerG']).sum().sort_values('Price', ascending = False).head(10)
```
