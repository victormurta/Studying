```python
#PLOTING
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ans = sns.load_dataset('anscombe')
print(ans)
```

```python
ds1 = ans[ans['dataset'] == 'I']
#por padrão faz um gráfico com linhas
plt.plot(ds1['x'], ds1['y'])
#agora com pontos:
plt.plot(ds1['x'], ds1['y'], 'o')
#interessante, faz tudo dentro da célula
```

```python
#agora com pontos:
plt.plot(ds1['x'], ds1['y'], 'o')
```

```python
ds2 = ans[ans['dataset'] == 'II']
ds3 = ans[ans['dataset'] == 'III']
ds4 = ans[ans['dataset'] == 'IV']
```

```python
#é possível fazer mais de um gráfico em uma mesma moldura
#definir o tamanho da moldura (2,2) e a posição do gráfico
#na mesmas (2,2,1)
fig = plt.figure()
axesl = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
 
axesl.plot(ds1['x'], ds1['y'],'o')
axes2.plot(ds2['x'], ds2['y'],'o')
axes3.plot(ds3['x'], ds3['y'],'o')
axes4.plot(ds4['x'], ds4['y'],'o')

axesl.set_title("dataset_l")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
```

```python
#ainda na seaborn, só que agora outro dataset
tips = sns.load_dataset('tips')
print(tips)
```

```python
#univariável = histogramas que só tem uma variável
#histogramas são os mais comuns para se observar
#uma única variável

fig = plt.figure()
axesl = fig.add_subplot(1, 1, 1)
axesl.hist(tips['total_bill'],  bins=10)
axesl.set_title('Histogram of Total Bill')
axesl.set_xlabel('Frequency')
axesl.set_ylabel('Total Bill')
fig.show ()
```

```python
#bivariável = temos 2 variáveis
#melhor utilizar o scatter plot (gráfico de dispersão)
scatter_plot = plt.figure()
axesl = scatter_plot.add_subplot(1, 1, 1)
axesl.scatter(tips['total_bill'],  tips['tip'])
axesl.set_title('Scatterplot of Total Bill vs Tip')
axesl.set_xlabel('Total Bill')
axesl.set_ylabel('Tip') 
scatter_plot.show()
```

```python
#Box plot = diagrama de caixa
boxplot = plt.figure()
axesl = boxplot.add_subplot(1, 1, 1)
axesl.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips [tips ['sex']== 'Male']['tip']],
              labels=['Female',   'Male'])
axesl.set_xlabel('Sex')
axesl.set_ylabel('Tip')
axesl.set_title('Boxplot of Tips by Sex')
```

```python
#Colocar mais de 3 variáveis é complicado
#Tamanho as vezes não é visto, cores podem confundir
#mas são mais indicadas
def recode_sex(sex):
   if sex == 'Female':
      return 0
   else:
      return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)
scatter_plot = plt.figure()
axesl = scatter_plot.add_subplot(1, 1, 1)
axesl.scatter(x=tips['total_bill'],
              y=tips['tip'],
              s=tips['size'] * 10,
              c=tips['sex_color'],
              alpha=0.5)
axesl.set_title('Total Bill vs Tip colored by Sex and sized by Size')
axesl.set_xlabel('Total Bill')
axesl.set_ylabel('Tip')
scatter_plot.show()
```

```python
#matplot é a base de fazer gráficos em python
#seaborn é feita em cima dela e tenta desenvolver mais
#para além de ser integrada com a pandas

#ela já define e faz algumas coisas por padrão:
hist = sns.displot(tips['total_bill'], kde = True)
hist.set(title = 'Total Bill Histogram with Density Plot')
```

```python
#é possível alterar para fazer utilizando outros padrões
hist = sns.displot(tips['total_bill'],kde=False)
hist.set(title = 'Total Bill Histogram', 
         xlabel = 'Total Bill',
         ylabel = 'Frequency')
```

```python
#density plot = gráfico de densidade(?)
den = sns.displot(tips['total_bill'], kind='kde')
den.set(title = 'Total Bill Density',
        xlabel = 'Total Bill',
        ylabel = 'Unit Probability')
```

```python
#rug plot = gráfico de tapete
#uma única variável quantitativa
 
hist_den_rug = sns.displot(tips['total_bill'], rug=True, kde = True)
hist_den_rug.set(title = 'Total Bill Histogram with Density and Rug Plot',
                xlabel = 'Total Bill')
```

```python
#Bar plot = Gráfico de barras
count = sns.countplot(x='day', data=tips)
count.set_title('Count of days')
count.set_xlabel('Day of the Week')
count.set_ylabel('Frequency')
```

```python
#agora seaborn com variáveis
#dispersão já calcula com linha de regressão

scatter = sns.regplot(x='total_bill',   y='tip',   data=tips)
scatter.set_title('Scatterplot of Total Bill and Tip')
scatter.set_xlabel('Total Bill')
scatter.set_ylabel('Tip')
```

```python
scatter = sns.jointplot(x='total_bill',   y='tip',  data=tips)
scatter.set_axis_labels(xlabel='Total Bill',   ylabel='Tip' )
scatter.fig.suptitle('Joint plot of Total Bill and Tip',
fontsize=20,   
y=1.03)
```

```python
#hexbin plot = é um de dispersão que, dependendo
#da dispersão e da quantidade dos dados, 
#facilita a visualização

hex = sns.jointplot(x="total_bill",  y="tip",data=tips,kind='hex')
hex.set_axis_labels(xlabel='Total Bill',   ylabel='Tip')
hex.fig.suptitle('Hexbin Joint plot of Total Bill and Tip',
                 fontsize=20,
                 y=1.03)
```

```python
#gráfico de densidade
kde = sns.kdeplot(x = tips['total_bill'],
                  y = tips['tip'],
                  fill = True)   
kde.set_title('Kernel Density Plot of Total Bill and Tip')
kde.set_xlabel('Total Bill')
kde.set_ylabel('Tip')
kde_joint = sns.jointplot(x='total_bill',   
                          y='tip',
                          data=tips,
                          kind='kde',
                         fill = True)
```

```python
#gráfico de barras
#IMPORTANTE: por padrão ele vai calcular a média mas da
#para alterar pelo parâmetro estimator

bar = sns.barplot(x='time',   y='total_bill' ,   data=tips)
bar.set_title('Barplot of average total bill for time of day')
bar.set_xlabel('Time of day')
bar.set_ylabel('Average total bill')
```

```python
#Gráfico de caixa
 
box = sns.boxplot(x='time',   y='total_bill',   data=tips)
box.set_title('Box plot of total bill by time of day')
box.set_xlabel('Time of day')
box.set_ylabel('Total Bill')
```

```python
#Gráfico de violino = violin plot
#bom complemento para o de caixa

violin = sns.violinplot(x='time', y='total_bill', data=tips)
violin.set_title('Violin plot of total bill by time of day')
violin.set_xlabel('Time of day')
violin.set_ylabel('Total Bill')
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.displot,rug=True)
```
